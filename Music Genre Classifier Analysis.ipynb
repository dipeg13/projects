```python
d = np.asarray(df)
```

```python
X = d[:,1:15]
y = d[:,15]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=.25, random_state=43)
```

```python
plt.hist(ytest)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

```python
pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('lr', LogisticRegression())])
```

```python
pipe.fit(Xtrain, ytrain)
```

```python
pipe.score(Xtest, ytest)
```

```python
from sklearn.metrics import plot_confusion_matrix as cm
```

```python
cm(pipe, Xtest, ytest)
```

```python
C = np.linspace(.1, 20.1, 200)
trainacc = []
testacc = []
for c in C:
    pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('lr', LogisticRegression(C=c, max_iter=2000))]).fit(Xtrain, ytrain)
    trainacc.append(pipe.score(Xtrain, ytrain))
    testacc.append(pipe.score(Xtest, ytest))
```

```python
plt.plot(C, trainacc, label='Train Accuracy')
plt.plot(C, testacc, label='Test Accuracy')
plt.legend()
plt.xlabel('C Parameter')
plt.ylabel('Accuracy')
plt.show()
```

```python
pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('lr', LogisticRegression(C=2.5, max_iter=2000))]).fit(Xtrain, ytrain)
cm(pipe, Xtest, ytest)
```

```python
print(mcm(ytest, pipe.predict(Xtest)))
```

We see that test accuracy converges to 55%, so C=18 would be a good choice of it.

```python
from sklearn.svm import SVC
```

```python
pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('svc', SVC(gamma=5))]).fit(Xtrain, ytrain)
```

```python
pipe.score(Xtest, ytest)
```

```python
C = np.linspace(.1, 20.1, 200)
trainacc = []
testacc = []
for c in C:
    pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('svc', SVC(gamma=c, max_iter=2000))]).fit(Xtrain, ytrain)
    trainacc.append(pipe.score(Xtrain, ytrain))
    testacc.append(pipe.score(Xtest, ytest))
```

```python
plt.plot(C, trainacc, label='Train Accuracy')
plt.plot(C, testacc, label='Test Accuracy')
plt.legend()
plt.xlabel('Gamma Parameter')
plt.ylabel('Accuracy')
plt.title('RBF')
plt.show()
```

We see that test accuracy converges to 60%, and with gamma>2, overfitting is begining to come up.

```python
C = np.linspace(.1, 20.1, 200)
trainacc = []
testacc = []
for c in C:
    pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('svc', SVC(C=c, gamma=2, max_iter=2000))]).fit(Xtrain, ytrain)
    trainacc.append(pipe.score(Xtrain, ytrain))
    testacc.append(pipe.score(Xtest, ytest))
```

```python
plt.plot(C, trainacc, label='Train Accuracy')
plt.plot(C, testacc, label='Test Accuracy')
plt.legend()
plt.xlabel('C Parameter')
plt.ylabel('Accuracy')
plt.title('RBF - Gamma=2')
plt.show()
```

```python
C = np.linspace(.1, 10.1, 100)
trainacc = []
testacc = []
for c in C:
    pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('svc', SVC(kernel='sigmoid', gamma=c, max_iter=2000))]).fit(Xtrain, ytrain)
    trainacc.append(pipe.score(Xtrain, ytrain))
    testacc.append(pipe.score(Xtest, ytest))
```

```python
plt.plot(C, trainacc, label='Train Accuracy')
plt.plot(C, testacc, label='Test Accuracy')
plt.legend()
plt.xlabel('Gamma Parameter')
plt.ylabel('Accuracy')
plt.title('Sigmoid')
plt.show()
```

Sigmoid kernel is rejected.

```python
C = np.linspace(.1, 10.1, 100)
trainacc = []
testacc = []
for c in C:
    pipe = Pipeline(steps=[('sc', StandardScaler()), ('mm', MinMaxScaler()), ('pca', PCA()), ('svc', SVC(kernel='sigmoid', gamma=.1, C=c, max_iter=2000))]).fit(Xtrain, ytrain)
    trainacc.append(pipe.score(Xtrain, ytrain))
    testacc.append(pipe.score(Xtest, ytest))
```

```python
plt.plot(C, trainacc, label='Train Accuracy')
plt.plot(C, testacc, label='Test Accuracy')
plt.legend()
plt.xlabel('C Parameter')
plt.ylabel('Accuracy')
plt.title('Sigmoid - Gamma=0.1')
plt.show()
```
